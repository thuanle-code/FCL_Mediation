```python
import pandas as pd
import numpy as np
import re
import os

FILE_XLSX = "Minh Tu_data.xlsx"

df_raw = pd.read_excel(FILE_XLSX, sheet_name=0)
df = df_raw.copy()
df.columns = [str(c).strip() for c in df.columns]

PREFIXES = ("RE", "EMP", "DOC", "CG", "SAT")

item_cols = [c for c in df.columns if any(str(c).startswith(p) for p in PREFIXES)]
df[item_cols] = df[item_cols].apply(pd.to_numeric, errors="coerce")

df_items = df[item_cols].dropna().reset_index(drop=True)

print("Số mẫu hợp lệ:", len(df_items))
print("Các biến:", df_items.columns.tolist())
```

```python
# ============================================================
# CRONBACH'S ALPHA – MINH TÚ
# ============================================================

def cronbach_alpha(items: pd.DataFrame) -> float:
    items = items.dropna()
    k = items.shape[1]
    if k < 2:
        return np.nan
    variances = items.var(ddof=1)
    total = items.sum(axis=1)
    total_var = total.var(ddof=1)
    if total_var == 0 or np.isnan(total_var):
        return np.nan
    return (k / (k - 1)) * (1 - variances.sum() / total_var)

def corrected_item_total_corr(items: pd.DataFrame) -> pd.Series:
    items = items.dropna()
    if items.shape[1] < 2:
        return pd.Series(dtype=float)
    total = items.sum(axis=1)
    return pd.Series({c: items[c].corr(total - items[c]) for c in items.columns})

def alpha_if_item_deleted(items: pd.DataFrame) -> pd.Series:
    items = items.dropna()
    out = {}
    for c in items.columns:
        out[c] = cronbach_alpha(items.drop(columns=[c]))
    return pd.Series(out)

def get_block(df_items: pd.DataFrame, prefix: str) -> pd.DataFrame:
    cols = [c for c in df_items.columns if str(c).startswith(prefix)]
    if len(cols) < 2:
        raise ValueError(f"Không tìm thấy đủ biến cho thang đo {prefix}.")
    return df_items[cols]

def cronbach_report(items: pd.DataFrame, scale_name: str, itc_threshold: float = 0.30):
    items = items.dropna()
    alpha_total = cronbach_alpha(items)
    itc = corrected_item_total_corr(items)
    alpha_del = alpha_if_item_deleted(items)

    rep = pd.DataFrame({
        "Biến quan sát": items.columns,
        "Corrected Item-Total Correlation": itc.values,
        "Cronbach's Alpha if Item Deleted": alpha_del.values
    })

    rep["Kết luận biến"] = np.where(
        rep["Corrected Item-Total Correlation"] < itc_threshold,
        "Loại",
        "Giữ"
    )

    rep = rep.sort_values("Corrected Item-Total Correlation").reset_index(drop=True)

    summary = {
        "Thang đo": scale_name,
        "Cronbach's Alpha": round(alpha_total, 3),
        "Số biến quan sát": items.shape[1],
        "Cỡ mẫu": items.shape[0],
        "Kết luận thang đo": "Đạt" if alpha_total >= 0.7 else "Chưa đạt"
    }

    print(f"\n===== {scale_name} =====")
    print(f"Cronbach's Alpha (overall) = {alpha_total:.3f} | k = {items.shape[1]} | n = {items.shape[0]}")
    print(rep.round(3).to_string(index=False))

    return summary, rep

scales = ["RE", "EMP", "DOC", "CG", "SAT"]

cronbach_summaries = []
cronbach_details = {}

for sc in scales:
    block = get_block(df_items, sc)
    summary, detail = cronbach_report(block, sc)
    cronbach_summaries.append(summary)
    cronbach_details[sc] = detail

summary_df = pd.DataFrame(cronbach_summaries)
summary_df.to_excel("MinhTu_Cronbach_Summary.xlsx", index=False)

with pd.ExcelWriter("MinhTu_Cronbach_Detail.xlsx", engine="openpyxl") as writer:
    for sc, detail in cronbach_details.items():
        detail.to_excel(writer, sheet_name=sc, index=False)

print("\n✅ Đã xuất:")
print("- MinhTu_Cronbach_Summary.xlsx")
print("- MinhTu_Cronbach_Detail.xlsx")
```

```python
# ============================================================
# EFA – MINH TÚ
# ============================================================

from scipy import stats
from sklearn.decomposition import PCA

N_FACTORS = 5
MIN_LOADING = 0.50

X = df_items.values
n, p = X.shape

print("EFA data shape (n, p) =", (n, p))
print("Items:", df_items.columns.tolist())

Xz = (X - X.mean(axis=0)) / X.std(axis=0, ddof=1)

def corr_matrix(xz: np.ndarray) -> np.ndarray:
    return np.corrcoef(xz, rowvar=False)

def kmo_test(R: np.ndarray):
    invR = np.linalg.inv(R)
    D = np.diag(1 / np.sqrt(np.diag(invR)))
    partial = -D @ invR @ D
    np.fill_diagonal(partial, 0.0)

    R2 = R.copy()
    np.fill_diagonal(R2, 0.0)

    r2 = (R2**2).sum()
    p2 = (partial**2).sum()
    kmo_overall = r2 / (r2 + p2)

    r2_j = (R2**2).sum(axis=0)
    p2_j = (partial**2).sum(axis=0)
    kmo_j = r2_j / (r2_j + p2_j)
    return kmo_overall, kmo_j

def bartlett_sphericity(R: np.ndarray, n_samples: int, p_vars: int):
    detR = np.linalg.det(R)
    detR = max(detR, 1e-30)
    chi2 = -(n_samples - 1 - (2*p_vars + 5)/6) * np.log(detR)
    df_ = p_vars*(p_vars-1)/2
    pval = 1 - stats.chi2.cdf(chi2, df_)
    return chi2, df_, pval

def varimax(Phi, gamma=1.0, q=50, tol=1e-6):
    p_, k_ = Phi.shape
    Rm = np.eye(k_)
    d = 0
    for _ in range(q):
        d_old = d
        Lambda = Phi @ Rm
        u, s, vh = np.linalg.svd(
            Phi.T @ (Lambda**3 - (gamma/p_) * Lambda @ np.diag(np.diag(Lambda.T @ Lambda)))
        )
        Rm = u @ vh
        d = s.sum()
        if d_old != 0 and d/d_old < 1 + tol:
            break
    return Phi @ Rm

R = corr_matrix(Xz)
kmo_overall, kmo_per_item = kmo_test(R)
chi2, df_b, p_b = bartlett_sphericity(R, n, p)

kmo_bartlett_df = pd.DataFrame({
    "Statistic": ["KMO Overall", "Bartlett Chi-square", "Bartlett df", "Bartlett p-value"],
    "Value": [kmo_overall, chi2, df_b, p_b]
})

kmo_item_df = pd.DataFrame({
    "Item": df_items.columns,
    "KMO_item": kmo_per_item
}).sort_values("KMO_item")

pca_all = PCA(n_components=p).fit(Xz)
eigenvalues = pca_all.explained_variance_
variance_ratio = pca_all.explained_variance_ratio_ * 100
cum_var = np.cumsum(variance_ratio)

eigen_table = pd.DataFrame({
    "Factor": np.arange(1, p+1),
    "Eigenvalue": eigenvalues,
    "% of Variance": variance_ratio,
    "Cumulative %": cum_var
})

pca = PCA(n_components=N_FACTORS).fit(Xz)
unrot_loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
rot_loadings = varimax(unrot_loadings)

loadings_df = pd.DataFrame(
    rot_loadings,
    index=df_items.columns,
    columns=[f"F{i+1}" for i in range(N_FACTORS)]
)

communalities = (rot_loadings**2).sum(axis=1)
comm_df = pd.DataFrame({
    "Item": df_items.columns,
    "Communality": communalities
}).sort_values("Communality")

absL = np.abs(rot_loadings)
top1 = absL.max(axis=1)
top2 = np.partition(absL, -2, axis=1)[:, -2] if N_FACTORS >= 2 else np.zeros_like(top1)
gap = top1 - top2

quality_df = pd.DataFrame({
    "Item": df_items.columns,
    "Top1_loading": top1,
    "Top2_loading": top2,
    "Gap": gap,
    f"Weak_loading(<{MIN_LOADING})": top1 < MIN_LOADING,
    "Cross_loading(Gap<0.20)": gap < 0.20
}).sort_values(
    [f"Weak_loading(<{MIN_LOADING})", "Cross_loading(Gap<0.20)", "Top1_loading"],
    ascending=[False, False, True]
)

print("\n================= KMO & BARTLETT =================")
print(kmo_bartlett_df.round(6).to_string(index=False))

print("\n================= EIGENVALUES =================")
print(eigen_table.round(4).to_string(index=False))

print("\n================= FACTOR LOADINGS (VARIMAX) =================")
print(loadings_df.round(3).to_string())

with pd.ExcelWriter("MinhTu_EFA_Summary.xlsx", engine="openpyxl") as writer:
    kmo_bartlett_df.to_excel(writer, sheet_name="KMO_Bartlett", index=False)
    kmo_item_df.to_excel(writer, sheet_name="KMO_per_item", index=False)
    eigen_table.to_excel(writer, sheet_name="Eigenvalues", index=False)
    loadings_df.round(3).to_excel(writer, sheet_name="Factor_Loadings")
    comm_df.round(3).to_excel(writer, sheet_name="Communalities", index=False)
    quality_df.round(3).to_excel(writer, sheet_name="Quality_Check", index=False)

print("\n✅ Đã xuất: MinhTu_EFA_Summary.xlsx")
```

```python
# ============================================================
# REGRESSION (OLS) – MINH TÚ
# SAT ~ RE + EMP + DOC + CG
# ============================================================

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.diagnostic import het_breuschpagan

def scale_mean(prefix):
    cols = [c for c in df_items.columns if str(c).startswith(prefix)]
    return df_items[cols].mean(axis=1)

df_model = pd.DataFrame({
    "RE":  scale_mean("RE"),
    "EMP": scale_mean("EMP"),
    "DOC": scale_mean("DOC"),
    "CG":  scale_mean("CG"),
    "SAT": scale_mean("SAT"),
})

print("Composite head:")
print(df_model.head())

y = df_model["SAT"]
X = df_model[["RE", "EMP", "DOC", "CG"]]
X_const = sm.add_constant(X)

model = sm.OLS(y, X_const).fit()

coef_table = pd.DataFrame({
    "B": model.params,
    "Std_Error": model.bse,
    "t": model.tvalues,
    "Sig": model.pvalues,
    "CI_Lower": model.conf_int()[0],
    "CI_Upper": model.conf_int()[1],
})

fit_table = pd.DataFrame({
    "N": [len(df_model)],
    "R_squared": [model.rsquared],
    "Adj_R_squared": [model.rsquared_adj],
    "F": [model.fvalue],
    "Sig_F": [model.f_pvalue],
    "AIC": [model.aic],
    "BIC": [model.bic],
    "Durbin_Watson": [sm.stats.stattools.durbin_watson(model.resid)],
})

vif_df = pd.DataFrame({
    "Variable": X_const.columns,
    "VIF": [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]
})

simple_rows = []
for var in ["RE", "EMP", "DOC", "CG"]:
    X1 = sm.add_constant(df_model[[var]])
    m1 = sm.OLS(y, X1).fit()
    simple_rows.append({
        "Model": f"SAT ~ {var}",
        "B": m1.params[var],
        "Sig": m1.pvalues[var],
        "R_squared": m1.rsquared,
        "Adj_R_squared": m1.rsquared_adj
    })

simple_df = pd.DataFrame(simple_rows)

jb_stat, jb_p, skew, kurt = sm.stats.stattools.jarque_bera(model.resid)
bp_stat, bp_p, _, _ = het_breuschpagan(model.resid, X_const)

assump_df = pd.DataFrame({
    "Jarque_Bera": [jb_stat],
    "Sig_JB": [jb_p],
    "Skewness": [skew],
    "Kurtosis": [kurt],
    "Breusch_Pagan": [bp_stat],
    "Sig_BP": [bp_p],
})

print("\n================= MODEL FIT =================")
print(fit_table.round(6).to_string(index=False))

print("\n================= COEFFICIENTS =================")
print(coef_table.round(6).to_string())

print("\n================= VIF =================")
print(vif_df.round(3).to_string(index=False))

with pd.ExcelWriter("MinhTu_Regression_Summary.xlsx", engine="openpyxl") as writer:
    df_model.round(3).to_excel(writer, sheet_name="Composite_Scores", index=False)
    fit_table.round(6).to_excel(writer, sheet_name="Model_Fit", index=False)
    coef_table.round(6).to_excel(writer, sheet_name="Coefficients")
    vif_df.round(3).to_excel(writer, sheet_name="VIF", index=False)
    simple_df.round(6).to_excel(writer, sheet_name="Simple_Regression", index=False)
    assump_df.round(6).to_excel(writer, sheet_name="Assumption_Checks", index=False)

print("\n✅ Đã xuất: MinhTu_Regression_Summary.xlsx")
```

```python
# ============================================================
# MEDIATION ANALYSIS – MINH TÚ
# CG đóng vai trò trung gian
# Chạy:
#   RE  -> CG -> SAT
#   EMP -> CG -> SAT
#   DOC -> CG -> SAT
# Xuất Excel: MinhTu_Mediation_Summary.xlsx
# ============================================================

import numpy as np
import pandas as pd
import statsmodels.api as sm

BOOT = 5000
SEED = 42
np.random.seed(SEED)

# ------------------------------------------------------------
# 1) Tạo lại df_model từ df_items
# ------------------------------------------------------------
def scale_mean(df_in, prefix):
    cols = [c for c in df_in.columns if str(c).startswith(prefix)]
    if len(cols) == 0:
        raise ValueError(f"Không tìm thấy biến nào cho prefix = {prefix}")
    return df_in[cols].mean(axis=1)

df_model = pd.DataFrame({
    "RE":  scale_mean(df_items, "RE"),
    "EMP": scale_mean(df_items, "EMP"),
    "DOC": scale_mean(df_items, "DOC"),
    "CG":  scale_mean(df_items, "CG"),
    "SAT": scale_mean(df_items, "SAT"),
}).dropna().reset_index(drop=True)

print("Valid n =", len(df_model))
print("Biến trong mô hình:", df_model.columns.tolist())
print("\nComposite head:")
print(df_model.head())

# ------------------------------------------------------------
# 2) Hàm OLS
# ------------------------------------------------------------
def ols_fit(df_in, y, X_list):
    X = sm.add_constant(df_in[X_list], has_constant="add")
    return sm.OLS(df_in[y], X).fit()

# ------------------------------------------------------------
# 3) Hàm trung gian đơn với bootstrap
# ------------------------------------------------------------
def mediation_single_boot(df_in, x, m, y, boot=5000, seed=42):
    rng = np.random.default_rng(seed)

    # a path: m ~ x
    mod_a = ols_fit(df_in, m, [x])
    a = mod_a.params[x]

    # c total: y ~ x
    mod_c = ols_fit(df_in, y, [x])
    c_total = mod_c.params[x]

    # b path + c' direct: y ~ x + m
    mod_b = ols_fit(df_in, y, [x, m])
    b = mod_b.params[m]
    c_prime = mod_b.params[x]

    indirect = a * b

    # bootstrap indirect
    n = len(df_in)
    boot_indirect = np.empty(boot, dtype=float)

    for i in range(boot):
        idx = rng.integers(0, n, n)
        samp = df_in.iloc[idx].reset_index(drop=True)

        try:
            a_b = ols_fit(samp, m, [x]).params[x]
            b_b = ols_fit(samp, y, [x, m]).params[m]
            boot_indirect[i] = a_b * b_b
        except Exception:
            boot_indirect[i] = np.nan

    boot_indirect = boot_indirect[~np.isnan(boot_indirect)]

    ci_low, ci_high = np.percentile(boot_indirect, [2.5, 97.5])
    sig = (ci_low > 0) or (ci_high < 0)

    res = {
        "a (m~x)": a,
        "b (y~x+m, coef m)": b,
        "c total (y~x)": c_total,
        "c' direct (y~x+m, coef x)": c_prime,
        "indirect a*b": indirect,
        "boot_CI_low_2.5%": ci_low,
        "boot_CI_high_97.5%": ci_high,
        "indirect_sig(CI excludes 0)": sig,
        "p_a(x)": mod_a.pvalues[x],
        "p_b(m)": mod_b.pvalues[m],
        "p_c_total(x)": mod_c.pvalues[x],
        "p_c_prime(x)": mod_b.pvalues[x],
    }

    return res, mod_a, mod_b, mod_c

# ------------------------------------------------------------
# 4) Chạy trung gian cho RE, EMP, DOC qua CG đến SAT
# ------------------------------------------------------------
mediation_results = {}

for x in ["RE", "EMP", "DOC"]:
    res, mod_a, mod_b, mod_c = mediation_single_boot(
        df_model, x=x, m="CG", y="SAT", boot=BOOT, seed=SEED
    )

    result_df = (
        pd.DataFrame.from_dict(res, orient="index", columns=["Value"])
        .reset_index()
        .rename(columns={"index": "Metric"})
    )

    # làm đẹp cột Value
    result_df["Value_num"] = pd.to_numeric(result_df["Value"], errors="coerce")
    result_df.loc[result_df["Value_num"].notna(), "Value"] = (
        result_df.loc[result_df["Value_num"].notna(), "Value_num"].round(6)
    )
    result_df = result_df.drop(columns=["Value_num"])

    mediation_results[f"{x}_CG_SAT"] = result_df

    print(f"\n================= MEDIATION: {x} -> CG -> SAT =================")
    print(result_df.to_string(index=False))

# ------------------------------------------------------------
# 5) Xuất Excel
# ------------------------------------------------------------
output_file = "MinhTu_Mediation_Summary.xlsx"

with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    for sheet_name, result_df in mediation_results.items():
        result_df.to_excel(writer, sheet_name=sheet_name[:31], index=False)

print(f"\n✅ Đã xuất file: {output_file}")
```

```python
# ============================================================
# SEQUENTIAL MEDIATION ANALYSIS – MINH TÚ
# Model 1: RE → EMP → CG → SAT
# Model 2: RE → DOC → CG → SAT
# ============================================================

import numpy as np
import pandas as pd
import statsmodels.api as sm

BOOT = 5000
SEED = 42
np.random.seed(SEED)

# ------------------------------------------------------------
# 1. LOAD DATA
# ------------------------------------------------------------

file_path = "Minh Tu_data.xlsx"

df_items = pd.read_excel(file_path)

print("Data loaded")
print("Columns:", df_items.columns.tolist())
print("Sample size:", len(df_items))

# ------------------------------------------------------------
# 2. Create composite variables
# ------------------------------------------------------------

def scale_mean(df_in, prefix):

    cols = [c for c in df_in.columns if str(c).startswith(prefix)]

    if len(cols) == 0:
        raise ValueError(f"No columns found for prefix {prefix}")

    return df_in[cols].mean(axis=1)


df_model = pd.DataFrame({

    "RE":  scale_mean(df_items,"RE"),
    "EMP": scale_mean(df_items,"EMP"),
    "DOC": scale_mean(df_items,"DOC"),
    "CG":  scale_mean(df_items,"CG"),
    "SAT": scale_mean(df_items,"SAT")

}).dropna().reset_index(drop=True)

print("\nComposite variables created")
print(df_model.head())

# ------------------------------------------------------------
# OLS helper
# ------------------------------------------------------------

def ols_fit(df,y,X):

    X = sm.add_constant(df[X])

    model = sm.OLS(df[y],X).fit()

    return model

# ------------------------------------------------------------
# Sequential mediation function
# ------------------------------------------------------------

def sequential_mediation(df,x,m1,m2,y,boot=5000):

    rng = np.random.default_rng(SEED)

    mod_a1 = ols_fit(df,m1,[x])
    a1 = mod_a1.params[x]

    mod_a2 = ols_fit(df,m2,[x,m1])
    a2 = mod_a2.params[m1]

    mod_b = ols_fit(df,y,[x,m1,m2])
    b = mod_b.params[m2]

    mod_c = ols_fit(df,y,[x])
    c_total = mod_c.params[x]

    mod_direct = ols_fit(df,y,[x,m1,m2])
    c_prime = mod_direct.params[x]

    indirect = a1*a2*b

    n = len(df)
    boot_vals = []

    for i in range(boot):

        idx = rng.integers(0,n,n)

        samp = df.iloc[idx]

        try:

            a1_b = ols_fit(samp,m1,[x]).params[x]

            a2_b = ols_fit(samp,m2,[x,m1]).params[m1]

            b_b = ols_fit(samp,y,[x,m1,m2]).params[m2]

            boot_vals.append(a1_b*a2_b*b_b)

        except:
            pass

    boot_vals = np.array(boot_vals)

    ci_low,ci_high = np.percentile(boot_vals,[2.5,97.5])

    sig = (ci_low>0) or (ci_high<0)

    result = {

        "a1 (M1~X)":a1,
        "a2 (M2~M1)":a2,
        "b (Y~M2)":b,

        "c_total":c_total,

        "c_prime":c_prime,

        "indirect_effect":indirect,

        "boot_CI_low":ci_low,

        "boot_CI_high":ci_high,

        "significant":sig

    }

    return result

# ------------------------------------------------------------
# Run models
# ------------------------------------------------------------

results = {}

print("\nRunning Model 1: RE → EMP → CG → SAT")

results["RE_EMP_CG_SAT"] = sequential_mediation(
    df_model,"RE","EMP","CG","SAT",BOOT
)

print("\nRunning Model 2: RE → DOC → CG → SAT")

results["RE_DOC_CG_SAT"] = sequential_mediation(
    df_model,"RE","DOC","CG","SAT",BOOT
)

# ------------------------------------------------------------
# Export Excel
# ------------------------------------------------------------

output_file = "MinhTu_Sequential_Mediation.xlsx"

with pd.ExcelWriter(output_file) as writer:

    for name,res in results.items():

        df_out = pd.DataFrame.from_dict(res,orient="index",columns=["Value"])

        df_out.to_excel(writer,sheet_name=name)

print("\nSequential mediation completed")

print("Results saved to:",output_file)
```

```python
# ============================================================
# SEM ANALYSIS – MINH TÚ
# ============================================================

import pandas as pd
from semopy import Model
from semopy import calc_stats

# ------------------------------------------------------------
# Load data
# ------------------------------------------------------------

df = pd.read_excel("Minh Tu_data.xlsx")

print("Sample size:", len(df))

# ------------------------------------------------------------
# Create composite variables
# ------------------------------------------------------------

def scale_mean(df, prefix):
    
    cols = [c for c in df.columns if str(c).startswith(prefix)]
    
    return df[cols].mean(axis=1)


df_model = pd.DataFrame({

    "RE": scale_mean(df,"RE"),
    "EMP": scale_mean(df,"EMP"),
    "DOC": scale_mean(df,"DOC"),
    "CG": scale_mean(df,"CG"),
    "SAT": scale_mean(df,"SAT")

})

print(df_model.head())

# ------------------------------------------------------------
# SEM model
# ------------------------------------------------------------

model_desc = """

EMP ~ RE
DOC ~ RE

CG ~ EMP + DOC

SAT ~ RE + EMP + DOC + CG

"""

# ------------------------------------------------------------
# Fit SEM
# ------------------------------------------------------------

model = Model(model_desc)

model.fit(df_model)

# ------------------------------------------------------------
# Path results
# ------------------------------------------------------------

results = model.inspect()

print("\nSEM PATH RESULTS")
print(results)

# ------------------------------------------------------------
# Model fit
# ------------------------------------------------------------

stats = calc_stats(model)

print("\nMODEL FIT")
print(stats)

# ------------------------------------------------------------
# Export results
# ------------------------------------------------------------

results.to_excel("MinhTu_SEM_results.xlsx")

print("\nSEM results saved to Excel")
```

```python
import pandas as pd
from semopy import Model, calc_stats

# =========================
# Load dataset
# =========================

df = pd.read_excel("Minh Tu_data.xlsx")

# =========================
# CFA MODEL
# =========================

model_desc = """

RE =~ RE1 + RE2 + RE3 + RE4 + RE5
EMP =~ EMP1 + EMP2 + EMP3 + EMP4 + EMP5
DOC =~ DOC1 + DOC2 + DOC3 + DOC4 + DOC5
CG =~ CG1 + CG2 + CG3 + CG4 + CG5
SAT =~ SAT1 + SAT2 + SAT3 + SAT4 + SAT5

"""

# =========================
# Fit model
# =========================

model = Model(model_desc)
model.fit(df)

# =========================
# Factor Loadings
# =========================

estimates = model.inspect()

factor_loadings = estimates[estimates["op"] == "~"]

# =========================
# Model Fit
# =========================

stats = calc_stats(model)

fit_indices = stats.T

# =========================
# Export to Excel
# =========================

with pd.ExcelWriter("CFA_results.xlsx") as writer:

    factor_loadings.to_excel(writer, sheet_name="Factor_Loadings", index=False)

    fit_indices.to_excel(writer, sheet_name="Model_Fit")

    estimates.to_excel(writer, sheet_name="All_Parameters", index=False)

print("CFA Results exported to CFA_results.xlsx")
```

```python
import pandas as pd
import statsmodels.formula.api as smf

# ===============================
# Load dataset
# ===============================

df = pd.read_excel("Minh Tu_data.xlsx")

# ===============================
# Create composite variables
# ===============================

df["RE"] = df[["RE1","RE2","RE3","RE4","RE5"]].mean(axis=1)
df["EMP"] = df[["EMP1","EMP2","EMP3","EMP4","EMP5"]].mean(axis=1)
df["DOC"] = df[["DOC1","DOC2","DOC3","DOC4","DOC5"]].mean(axis=1)
df["CG"] = df[["CG1","CG2","CG3","CG4","CG5"]].mean(axis=1)
df["SAT"] = df[["SAT1","SAT2","SAT3","SAT4","SAT5"]].mean(axis=1)

# ===============================
# Mean centering
# ===============================

vars_list = ["RE","EMP","DOC","CG","SAT"]

for v in vars_list:
    df[v+"_c"] = df[v] - df[v].mean()

# ===============================
# Create interaction terms
# ===============================

df["RE_EMP"] = df["RE_c"] * df["EMP_c"]
df["RE_DOC"] = df["RE_c"] * df["DOC_c"]
df["EMP_CG"] = df["EMP_c"] * df["CG_c"]
df["DOC_CG"] = df["DOC_c"] * df["CG_c"]
df["CG_SAT"] = df["CG_c"] * df["SAT_c"]

# ===============================
# Moderation regressions
# ===============================

results = {}

model1 = smf.ols("EMP ~ RE_c + DOC_c + RE_EMP", data=df).fit()
results["RE_EMP_mod"] = model1.summary2().tables[1]

model2 = smf.ols("DOC ~ RE_c + EMP_c + RE_DOC", data=df).fit()
results["RE_DOC_mod"] = model2.summary2().tables[1]

model3 = smf.ols("CG ~ EMP_c + DOC_c + EMP_CG", data=df).fit()
results["EMP_CG_mod"] = model3.summary2().tables[1]

model4 = smf.ols("CG ~ DOC_c + EMP_c + DOC_CG", data=df).fit()
results["DOC_CG_mod"] = model4.summary2().tables[1]

model5 = smf.ols("SAT ~ CG_c + EMP_c + CG_SAT", data=df).fit()
results["CG_SAT_mod"] = model5.summary2().tables[1]

# ===============================
# Export to Excel
# ===============================

with pd.ExcelWriter("Moderation_results.xlsx") as writer:

    for name, table in results.items():
        table.to_excel(writer, sheet_name=name)

print("Moderation results exported to Moderation_results.xlsx")
```
